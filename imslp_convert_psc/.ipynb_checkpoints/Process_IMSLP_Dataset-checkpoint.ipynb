```python
def isValidPDF(infile):
    try:
        PyPDF2.PdfFileReader(open(infile, "rb"))
        isValid = True
    except:
        isValid = False
    return isValid
```

```python
def parsePdfInfo(string):
    # an example pdf info string:
    # '*#400812 - 22.72MB, 8 pp. - 0.0/10\n\n2\n4\n6\n8\n10\n\n (-) - V/V/V - 2151×⇩ - Piupianissimo'
    string = string.replace('\n','')
    m = re.search('#(\d+)\s', string)
    pdf_id = m.group(1)
    m = re.search('-\s(\d+)×', string)
    if m:
        num_downloads = int(m.group(1))
    else:
        num_downloads = 0 # no downloads field
    return pdf_id, num_downloads
```

```python
def parseHtml(html_file):
    with open(html_file,'r') as f:
        text = f.read()
    soup = BeautifulSoup(text, 'html.parser')
    return soup
```

```python
def getField(d, key):
    if key in d:
        result = d[key]
    else:
        result = ''
    return result
```

```python
def extractScoreInfo(node):
    
    # read all table entries
    d = {}
    for row in node.find("table").find("table").findAll("tr"):
        row_header = row.find("th").text.strip().lower()
        row_field = row.find("td").text.strip().lower()
        d[row_header] = row_field
        #print('{}|{}'.format(row_header, row_field))
    
    # extract relevant fields
    publisherInfo = getField(d, 'publisher. info.')
    copyright = getField(d, 'copyright')
    
    # check if valid
    isPublicDomain = 'public domain' in copyright and 'non-pd' not in copyright
    isCreativeCommons = 'creative commons' in copyright
    isShareable = isPublicDomain # or isCreativeCommons
    isManuscript = 'manuscript' in publisherInfo
    
    return isShareable, isManuscript
```

```python
def selectSinglePDF(pieceDir):
    '''
    Selects a single PDF file within a piece directory.  First filter the list of pdfs to ensure that the piece
    has a suitable copyright and is not a manuscript, and then within the remainining options select the 
    most popular (i.e. most downloaded) score.  
    
    Returns the full path to the selected pdf file.  If no valid pdfs are found, returns None.
    '''    # get pdf ids, number of downloads
    html_file = '{}/html.txt'.format(pieceDir)
    soup = parseHtml(html_file)
    scores_list = soup.find(id='tabScore1')
    if scores_list is None: # no scores
        return None
    tuples = [] # populate with (pdf_id, num_downloads)
    for i, child in enumerate(scores_list.findAll("div", recursive=False)): # list of scores
        try:
            isShareable, isManuscript = extractScoreInfo(child) # copyright ok & not a manuscript
        except: # incorrectly formatted entry -- skip
            continue
        for pdf_div_tag in child.findAll("div", recursive=False): # may have multiple pdfs
            we_file_info2 = pdf_div_tag.find(class_='we_file_info2')
            if we_file_info2 is not None: # sometimes there are div tags with additional information
                pdf_id, num_downloads = parsePdfInfo(we_file_info2.text)
                tuples.append((pdf_id, num_downloads, isShareable, not isManuscript))

    # sort by downloads, verify that PDF can be read
    valid_list = []
    for (pdf_id, num_downloads, isPublicDomain, notManuscript) in tuples:
        fullpath = '{}/{}.pdf'.format(pieceDir, pdf_id)
        #print((pdf_id, num_downloads, isPublicDomain, notManuscript))
        if isPublicDomain and notManuscript and isValidPDF(fullpath):
            valid_list.append((fullpath, num_downloads, isPublicDomain, notManuscript))
    return valid_list
```

```python
filelist = '/home/dyang/imslp_convert/fileList1.txt'
imslp_basedir = '/data1/dbashir/Project/'
composers = []
composerPDFCount = {}
pdfCount = 0
pieceCount = 0

if os.path.exists(filelist):
    with open(filelist,'r') as f:
        for line in f:
            line = line[:-1]
            composer = line.split('/')[-2]
            piece = line.split('/')[-1]
            pieceDir = imslp_basedir+line
            if not composer in composers:
                composers.append(composer)
                composerPDFCount[composer] = 0
                
            for subdir, dirs, files in os.walk(pieceDir):
                for file in files:
                    if file[-4:]=='.pdf':
                        pdfCount+=1
                        composerPDFCount[composer] = composerPDFCount[composer]+1
            pieceCount+=1
    print("Number of PDFs: ", pdfCount)
    print("Unique Pieces: ", pieceCount)
    print("Unique Composers: ",len(composers))
    
    
```
