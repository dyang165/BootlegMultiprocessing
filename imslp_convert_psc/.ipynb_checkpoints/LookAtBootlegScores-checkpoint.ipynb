```python
from ExtractBootlegFeatures1 import *
import pickle
import numpy as np
```

```python
bscore_file = "/home/dyang/95021.pkl"
```

```python
def decodeColumn(num):
    col = []
    for i in range(62):
        col.insert(0,num%2)
        num = int(num/2)
    return col
```

```python
def getTotalBscore(bscore_file):
    bscore_array = []
    with open (bscore_file,'rb') as f:
        bscore_array = pickle.load(f)
    f.close()
    total_bscore = np.array([]).reshape(62,0)
    page_array = []
    for page in bscore_array:
        total_page = np.array([]).reshape(62,0)
        for num in page:
            col = np.array(decodeColumn(num)).reshape(62,-1)
            total_page = np.concatenate((total_page,col),axis=1)
        total_bscore = np.concatenate((total_bscore,total_page),axis=1)
        page_array.append(total_page)
    return total_bscore,page_array
```

```python
staffLinesBoth = [13,15,17,19,21,35,37,39,41,43]
visualizeLongBootlegScore(total_bscore, staffLinesBoth)
```

```python
pageNum = 4
visualizeLongBootlegScore(page_array[pageNum-1], staffLinesBoth)
```

```python
import os
import sys
bootleg_dir = "/data1/dyang/Datasets/imslp_bootleg_dir/"
outFile = "/home/dyang/imslp_convert_psc/FailedFiles.txt"

for cd, subdirs, files in os.walk(bootleg_dir):
    for file in files:
        fname = os.path.join(cd,file)
        bscore,_ = getTotalBscore(fname)
        if np.array(bscore).shape[1] == 0:
            print(fname,file=open(outFile,"a"))
            print(fname)
```
